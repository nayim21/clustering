# k-MEANS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
#matplotlib.use('Agg')
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn import metrics
import seaborn as sns
import seaborn as sns; sns.set()
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
df = pd.read_csv(r"D:\dataset\iris.csv")
x = df.iloc[:,[0,1,2,3]].values
```

```python
from sklearn.cluster import KMeans

wcss=[]
for i in range(1,11) :
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
               
plt.plot(range(1,11),wcss)
plt.show()
```

```python
i=3
kmeans=KMeans(n_clusters=i)
y_kmeans=kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100,c='red',label='s')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='blue',label='ve')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100,c='green',label='vi')
print(kmeans.cluster_centers_)

print("Num of iter to Converage:",kmeans.n_iter_)
#print("label",kmeans.labels_)
print(kmeans.inertia_)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='yellow',label='center')
plt.legend()
#plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='yellow',label='center')
#plt.legend()
```

# DBSCAN

```python
clusters = DBSCAN(eps=4, min_samples=100).fit(x)
# get cluster labels
clusters.labels_
p = sns.scatterplot(data=x )
```

# Hierarchical Clustering

```python
linkage_data = linkage(x, method='single', metric='euclidean')
dendrogram(linkage_data)
plt.show()
```
